Liniowa analiza dyskryminacyjna (ang. linear discriminant analysis, LDA) jest używana w uczeniu maszynowym do znalezienia liniowej kombinacji cech, które najlepiej rozróżniają dwie lub więcej klas obiektów lub zdarzeń. LDA ma na celu znalezienie liniowej kombinacji zmiennych (cech), która maksymalnie rozdziela klasy. Innymi słowy, LDA dąży do znalezienia projekcji danych wejściowych na niższowymiarową przestrzeń, w której klasy są jak najbardziej separowalne. Zastosowanie LDA można znaleźć np. w modelowaniu ryzyka kredytowego. Ryzyko kredytowe to ryzyko straty, jaką może ponieść jedna ze stron umowy, np. bank, udzielając kredytu osobie lub firmie, która nie będzie w stanie go spłacić zgodnie z warunkami umowy kredytowej. Jest to istotna wartość szczególnie dla instytucji bankowych w przypadku decyzji o udzielaniu kredytów. W ich interesie jest to, aby klient był jaknajbardziej wiarygodny, tj. z jak największym prawdopodobieństwem spłacił kredyt z jak najmniejszymi potencjalnymi komplikacjami. Modele empiryczne są jednym z narzędzi, które wspomagają analizę i zarządzanie ryzykiem kredytowym. Są to matematyczne modele oparte na danych empirycznych, które pomagają prognozować prawdopodobieństwo niewypłacalności kredytobiorcy. Modele te opierają się na statystycznej analizie historycznych danych np. wskaźnikach finansowych firm skałdających wnioski o kredyt. 

Na początek definiujemy modele, które zostają testowane w dalszej części projektu. 
Poniższe modele są narzędziami służącym do oceny ryzyka upadłości firm. Modele te starają się przewidzieć bankructwo firmy (lub ich brak) na podstawie kilku wskaźników finansowych, takich jak między innymi wskaźnik zadłużenia i rentowności. Z wyjątkiem modelu Fulmera, modele są liniowy, co niekoniecznie odzwierciedla drogę do bankructwa, a operują na danych księgowych (dane są dostępne w pewnych odstępach czasowych, np. kwartalne) oraz zawierają tylko kilka zmiennych, co niekoniecznie odzwierciedla realną sytuację firmy, co potiwerdzają skuteczności modeli na poziomie kilkudziesięciu procent poprawnych klasyfikacji. 


Model Altmana
X1 aktywa obrotowe / aktywa ogółem
X2 zysk zatrzymany / aktywa ogółem
X3 zysk przed spłatą odsetek i podatkiem / aktywa ogółem
X4 wartość rynkowa kapitału akcyjnego / wartość księgowa zadłużenia
X5 przychody ze sprzedaży / aktywa ogółem

Model Springate'a
X1 kapitał pracujący / aktywa ogółem
X2 zysk przed spłatą odsetek i podatkiem / aktywa ogółem
X3 wynik brutto / zobowiązania krótkoterminowe
X4 przychody ze sprzedaży / aktywa ogółem

Model Mączyńskiej
X1 zysk / aktywa ogółem
X2 kapital własny / aktywa ogółem
X3 zysk netto + amortyzacja / kapitaly obce
X4 aktywa obrotowe / zobowiazania krótkoterminowe
X5 przychody netto ze sprzedazy + pozostale/ aktywa ogółem

Model poznański
X1 wynik finansowy netto / aktywa ogółem 
X2 aktywa obrotowe ogółem / zobowiązania krótkoterminowe
X3 kapitał własny + kapytały obce długoterminowe / aktywa ogółem
X4 zysk brutto ze sprzedaży / przychody netto ze sprzedaży

Model Hadasik
X1 aktywa obrotowe ogółem / zobowiązania krótkoterminowe
X2 aktywa obrotowe ogółem - zapasy - rozliczenia międzyokresowe / zobowiązania krótkoterminowe
X3 kapitały obce / aktywa ogółem
X4 kapitał obrotowy netto / aktywa ogółem
X5 należności krótkoterminowe / przychody netto ze sprzedaży
X6 zapasy / przychody netto ze sprzedaży

Model Fulmera
X1 zysk zatrzymany / aktywa ogółem
X2 przychody ogółem / aktywa ogółem
X3 zysk brutto / kapitały własne
X4 przepływy pieniężne / zobowiązania ogółem
X5 zobowiązania ogółem / aktywa ogółem
X6 zobowiązania bieżące / aktywa ogółem
X7 log(aktywa rzeczowe)
X8 kapitał obrotowy / zobowiązania ogółem
X9 log(zysk przed podatkiem i odsetki)


Stworzenie wlasnego modelu
Na podstawie analizy zmiennych wykorzystywanych w innych modelach i ich współczynnikach, wielkości wartości zmiennych oraz metodzie prób i błędów przyjęto poniższe zmienne i ich współczynniki
Własny model:
X1 aktywa obrotowe / aktywa ogółem
X2 zobowiązania ogółem / aktywa ogółem  
X3 przychody ze sprzedaży / aktywa ogółem
X4 przepływy pieniężne / zobowiązania ogółem
X5 zysk brutto / kapitały własne

x = X1 * 1.5 - X2 * 0.7 + X3 * 0.6 + X4 * 0.25 + X5 * 0.45
gdzie 
x>=1 - brak upadłości firmy
x<1 - upadłość firmy 


```python
import math

def altman_model(x1,x2,x3,x4,x5):
    x1 = x1.replace(",",".")
    x2 = x2.replace(",",".")
    x3 = x3.replace(",",".")
    x4 = x4.replace(",",".")
    x5 = x5.replace(",",".")
    z = 1.2*float(x1)+1.4*float(x2)+3.3*float(x3)+0.6*float(x4)+0.99*float(x5)
    return z

def springate_model(x1,x2,x3,x4):
    x1 = x1.replace(",",".")
    x2 = x2.replace(",",".")
    x3 = x3.replace(",",".")
    x4 = x4.replace(",",".")
    z = 1.03*float(x1) + 3.07*float(x2) + 0.66*float(x3) + 0.4*float(x4)
    return z

def poznanski_model(x1,x2,x3,x4):
    x1 = x1.replace(",",".")
    x2 = x2.replace(",",".")
    x3 = x3.replace(",",".")
    x4 = x4.replace(",",".")
    z = 3.562*float(x1) + 1.58*float(x2) + 4.288*float(x3) + 6.719*float(x4) - 2.368
    return z

def maczynska_model(x1,x2,x3,x4,x5):
    x1 = x1.replace(",",".")
    x2 = x2.replace(",",".")
    x3 = x3.replace(",",".")
    x4 = x4.replace(",",".")
    x5 = x5.replace(",",".")
    z = 9.478*float(x1)+3.613*float(x2)+3.246*float(x3)+0.455*float(x4)+0.802*float(x5)-2.478
    return z

def hadasik_model(x1,x2,x3,x4,x5,x6):
    x1 = x1.replace(",",".")
    x2 = x2.replace(",",".")
    x3 = x3.replace(",",".")
    x4 = x4.replace(",",".")
    x5 = x5.replace(",",".")
    x6 = x6.replace(",",".")
    z = 0.365425*float(x1)-0.765526*float(x2)-2.40435*float(x3)+1.59079*float(x4)+0.00230258*float(x5)-0.0127827*float(x6)+2.36261
    return z

def fulmer_model(x1,x2,x3,x4,x5,x6,x7,x8,x9):
    x1 = x1.replace(",",".")
    x2 = x2.replace(",",".")
    x3 = x3.replace(",",".")
    x4 = x4.replace(",",".")
    x5 = x5.replace(",",".")
    x6 = x6.replace(",",".")
    x7 = x7.replace(",",".")
    x8 = x8.replace(",",".")
    x9 = x9.replace(",",".")
    z = 5.528*float(x1)+0.212*float(x2)+0.07*float(x3)+1.27*float(x4)
    -0.12*float(x5)+2.335*float(x6)+0.575*float(x7)+10.83*float(x8)+(0.894*math.log(float(x9)) if float(x9)>0 else 0)-6.075
    return z

def own_model(x1, x2, x3, x4, x5):
    x1 = x1.replace(",",".")
    x2 = x2.replace(",",".")
    x3 = x3.replace(",",".")
    x4 = x4.replace(",",".")
    x5 = x5.replace(",",".")
    z = 1.5*float(x1)-0.7*float(x2)+0.6*float(x3)+0.25*float(x4)+0.45*float(x5)
    return z


```

Mimo to wyższą precyzją mogą pochwalić się modele m.in poznański czy Hadasika, który z nich wszystkich jest najdokładniejszy, choć to model Springate'a ma największą czułość i najlepiej klasyfikował rzeczywistych bankrutów.

Ze względu na to, że liczebność klas bankrutów i niebankrutów jest nierówna (firm - bankrutów jest kilkanaście razy mniej), zauważalna jest duża różnica pomiędzy wartościami czułości i specyficzności w prawie każdym z modeli. Klas 0 - niebankrutów, jest znacznie więcej, dlatego łatwo jest założyć w ciemno i zaklasyfikować firmę do nie bankruta, a prawdopodobieństwo poprawności takiej klasyfikacji będzie wysokie co potiwerdzają powyższe wyniki. Natomiast ważniejsza jest umiejętność wyłapywania tych drugich przypadków - ważniejsze dla banku jest to, aby uniknąć klasyfikacji potencjalnego bankruta do niebankruta, niż klasyfikacji niebankruta do bankruta, ponieważ mniejszym problemem dla banku jest utracenie klienta (klasyfikacja niebankruta do bankruta) niż problemy z odzyskaniem należności (klasyfikacja bankruta to niebankruta).
